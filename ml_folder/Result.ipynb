```python
import matplotlib.pyplot as plt
import numpy as np

# Data for WA attack
labels = ['F^c', 'F^u', 'F^r']
models = ['CNN', 'RF', 'LR']
no_atk_recall = {
    'CNN': [0.95, 0.60, 0.81],
    'RF': [0.94, 0.54, 0.95],
    'LR': [0.87, 0.32, 0.79]
}
wa_u_recall = {
    'CNN': [0.90, 1.00, 0.80],
    'RF': [0.88, 1.00, 0.95],
    'LR': [0.58, 0.00, 0.79]
}
wa_r_recall = {
    'CNN': [0.96, 0.60, 0.35],
    'RF': [0.95, 0.54, 0.84],
    'LR': [0.78, 0.32, 0.60]
}
wa_c_recall = {
    'CNN': [0.91, 1.00, 0.34],
    'RF': [0.78, 1.00, 0.84],
    'LR': [0.44, 0.00, 0.59]
}

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, axs = plt.subplots(1, 3, figsize=(15, 6), sharey=True)

for i, model in enumerate(models):
    ax = axs[i]
    ax.bar(x - 1.5*width, no_atk_recall[model], width, label='no-atk')
    ax.bar(x - 0.5*width, wa_u_recall[model], width, label='WA^u')
    ax.bar(x + 0.5*width, wa_r_recall[model], width, label='WA^r')
    ax.bar(x + 1.5*width, wa_c_recall[model], width, label='WA^c')

    ax.set_title(model)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylim(0, 1.1)
    ax.set_xlabel('Feature')
    ax.set_ylabel('tpr')

# Add a single legend for the whole figure
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper left', bbox_to_anchor=(1, 1))

fig.tight_layout()
plt.show();
```
